```python
from selenium import webdriver
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
import time
import datetime
```

```python
def daum_crawling():
    driver = webdriver.Chrome()
    title_ls,link_ls,time_ls = [],[],[]

    for page in range(1,21+1):
        try : 
            url = 'http://m.finance.daum.net/m/news/news_list.daum?type=main&page={}'.format(page)
            driver.get(url)
            time.sleep(1)
            for content in range(1,10+1):
#                 if time_ls != [] :
#                     if time_ls != str(datetime.datetime.now().month) + str(datetime.datetime.now().day):
#                         break
                if content %5 == 0 : time.sleep(1)
                driver.execute_script('window.scrollTo(10,{});'.format(content*150))
                title_ls.append(driver.find_element_by_css_selector(\
            '#mArticle > div.all_list > ul > li:nth-child({}) > a'\
            .format(content)).text)
                link_ls.append(driver.find_element_by_css_selector(\
            '#mArticle > div.all_list > ul > li:nth-child({}) > a'\
            .format(content)).get_attribute('href'))

        except Exception as e :
            print(e)

    driver.close()
    
    df = pd.DataFrame()
    df['title'] = title_ls
    df['link'] = link_ls
#     df['time'] = time_ls
    
    return df
```

```python
df = daum_crawling()
```

```python
df
```

```python
def content_append_daum():
    driver = webdriver.Chrome()
    content_ls = []
    for i in range(len(df)):
        url = df['link'][i]
        driver.get(url)
        time.sleep(1)
        content_ls.append(driver.find_element_by_css_selector('#dmcfContents > section').text)
    driver.close()
    return content_ls
```

```python
content_ls = content_append_daum()

df['content'] = content_ls

df.tail()
```

```python
df.to_csv('{}_with_link_daum.csv'.format(str(datetime.datetime.now().month) + str(datetime.datetime.now().day),index=False),index=False)
```
